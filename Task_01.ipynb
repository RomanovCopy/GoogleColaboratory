```python
import pandas as pd
df=pd.read_csv('https://minobrnauki.gov.ru/upload/iblock/78d/78d77b235f7b3b14e1ac671e61435311.csv', sep=';' , encoding='cp1251')
maskacfo=df['Субъект РФ'].str.contains('Центральный федеральный округ')#маска для поиска индекса ЦФО
maskafo=df['Субъект РФ'].str.contains('федеральный округ')#маска для поиска индексов ФО
CFD_index =df[maskacfo].index#индекс ЦФО
Indices_of_federal_districts = [df[maskafo].index]#индексы федеральных округов
List_of_regions_of_the_CFD = df.loc[2:19]#список регионов ЦФО
Regions_of_the_CFD = List_of_regions_of_the_CFD[List_of_regions_of_the_CFD['Субъект РФ'].str.contains('область')]#Список областей ЦФО
max = Regions_of_the_CFD['Численность студентов очно-заочная (вечерняя) форма, человек, 2015'].max()#максимальное колличество студентов
result = Regions_of_the_CFD[Regions_of_the_CFD['Численность студентов очно-заочная (вечерняя) форма, человек, 2015']==max]#ячейка со значением max
result
```
